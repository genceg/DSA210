```python
h1 = pd.read_csv('Happiness2005-2023.csv', encoding='latin1')
h2 = pd.read_csv('Happiness2024.csv',     encoding='latin1')

h2['Year'] = 2024

h1.rename(columns={
    'Country name':'Country',
    'year':'Year',
    'Life Ladder':'HappinessScore'
}, inplace=True)
h2.rename(columns={
    'Country name':'Country',
    'Ladder score':'HappinessScore'
}, inplace=True)

h = pd.concat([h1,h2], ignore_index=True)
h = h.dropna(subset=['Country','Year','HappinessScore'])
h = h[(h.Year>=2014)&(h.Year<=2024)]

h.head()
```

```python
# ─── Cell 6: Load & clean HDI from HDIdata.csv ────────────
import pandas as pd

# 1) Read, skipping the first 5 rows so that row 6 becomes the header
hdi = pd.read_csv('HDIdata.csv', skiprows=5)

# 2) Check what columns pandas picked up
print("Columns after skiprows:", hdi.columns.tolist())

# 3) Select only the two you need
hdi = hdi[['Country', 'Value']]

# 4) Rename the HDI column for simplicity
hdi.rename(columns={'Value': 'HDI'}, inplace=True)

# 5) Drop any rows missing Country or HDI
hdi = hdi.dropna(subset=['Country', 'HDI'])

# 6) Quick preview
hdi.head()
```

```python
temp = pd.read_csv('average_temperatures.csv', encoding='latin1')
temp.rename(columns={'Country':'Country','Temperature_C':'AvgTemp'}, inplace=True)
temp = temp.dropna(subset=['Country','AvgTemp'])

ineq = pd.read_csv('income_inequality.csv', encoding='latin1')
ineq.rename(columns={
    'country_name':'Country',
    'year':'Year',
    'value':'GiniIndex'
}, inplace=True)
ineq = ineq.dropna(subset=['Country','Year','GiniIndex'])
ineq = ineq[(ineq.Year>=2014)&(ineq.Year<=2024)]

inet = pd.read_csv('useofInternet.csv', encoding='latin1')
inet.rename(columns={
    'Country or area':'Country',
    'Year':'Year',
    'Percentage_Internet_Users':'InternetPct'
}, inplace=True)
inet = inet.dropna(subset=['Country','Year','InternetPct'])
inet = inet[(inet.Year>=2014)&(inet.Year<=2024)]

act = pd.read_csv('PhysicalActivity.csv', encoding='latin1')
act.rename(columns={'Country':'Country'}, inplace=True)
act = act.dropna(subset=['Country'])

fifa = pd.read_csv('FifaRankings.csv', encoding='latin1')
fifa['Year'] = pd.to_datetime(fifa['rank_date']).dt.year
fifa.rename(columns={
    'country_full':'Country',
    'rank':'FifaRank'
}, inplace=True)
fifa = fifa.dropna(subset=['Country','Year','FifaRank'])
fifa = fifa[(fifa.Year>=2014)&(fifa.Year<=2024)]
```

```python
merged = h.merge(hdi, on='Country', how='inner')

for df in [ineq, inet, fifa]:
    merged = merged.merge(df, on=['Country','Year'], how='left')

merged = merged.merge(temp, on='Country', how='left')
merged = merged.merge(act,  on='Country', how='left')

merged.to_csv('merged_2014_2024.csv', index=False)
print("Merged shape:", merged.shape)
merged.head()
```
